# Creating a workflow Diagram

The package uses `graphviz` to create a tailored workflow model to help you explain the model you are using.  Here, we will show you how to produce it.

Please note that you will need to have preprocessed and created your model to load a workflow diagram.

```python
import pandas as pd
import os
import sys

# Get the path of the current script
current_script_path = os.path.abspath("__file__")

# Deduce the root folder of the project
project_root = os.path.dirname(os.path.dirname(os.path.dirname(current_script_path)))

# Add the project root to the Python path
sys.path.append(project_root)
```

```python
from datascribe.datasets import load_ed_scribe_evaluated
```

We will load in the Scribe object with the analysis performed so far (note: analysis graphs will appear).

```python
s = load_ed_scribe_evaluated()
```

To create an image ready for the output, simply use the `Scribe.workflow.create_workflow_image()` method below.  A preview will not appear but the image file will be stored in the chosen directory in a subfolder called **images**.

```python
s.workflow.create_workflow_image()
```

For the example, the workflow image saved looks like the below image:

![Workflow diagram of data processing](example_images/model%20workflow.png)

The steps in the image will vary depending on the preprocessing tools performed through the Scribe object.  When future models are included in the package, this part of the diagram would adapt to the input as well.
