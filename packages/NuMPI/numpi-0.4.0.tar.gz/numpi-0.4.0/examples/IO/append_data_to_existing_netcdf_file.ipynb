```python
from NuMPI.IO.NetCDF import NCStructuredGrid
import numpy as np
```

```python

n = 4

nc = NCStructuredGrid("data_1.nc", "w", (n,n))

nc[0].scalar1 = 0.2
nc[0].scalar2 =4

nc[0].field1 = np.random.normal(size=(n,n))
nc[0].field2 = np.random.normal(size=(n,n))

nc.close()
```

```python
!ncdump data_1.nc
```

Can I add some scalar data to it, afterwards ? 

```python

n = 4

nc = NCStructuredGrid("data_1.nc", "a", (n,n))

nc[0].scalar3 =4
nc[0].field3 = np.random.normal(size=(n,n))




nc.close()
```

```python
!ncdump data_1.nc
```

# What happens when I hav more then one frame ? 

```python

n = 4

nc = NCStructuredGrid("data_2.nc", "w", (n,n))

nc[0].scalar1 = 0.2
nc[0].scalar2 = 4

nc[0].field1 = np.random.normal(size=(n,n))
nc[0].field2 = np.random.normal(size=(n,n))


nc[1].scalar1 = 0.1
nc[1].scalar2 = 8

nc[1].field1 = np.random.normal(size=(n,n))
nc[1].field2 = np.random.normal(size=(n,n))


nc[2].scalar1 = 0.3
nc[2].scalar2 = 45

nc[2].field1 = np.random.normal(size=(n,n))
nc[2].field2 = np.random.normal(size=(n,n))




nc.close()


```

```python
!ncdump data_2.nc
```

```python

n = 4

nc = NCStructuredGrid("data_2.nc", "a", (n,n))

nc[1].scalar3 = 32154
nc[0].field3 = np.random.normal(size=(n,n))

nc.close()



```

```python
!ncdump data_2.nc
```
